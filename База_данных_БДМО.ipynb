```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Загружаю данные, создаю таблицы a,b,c без ненужными столбцов
BDMO_data_1 = pd.read_csv(r"/Users/boris/Downloads/БДМО/1 indicator clean v2.csv")
a = BDMO_data_1[["mun_type", "municipality", "oktmo", "oktmo_munr", "rayon", "year"]]

BDMO_data_2 = pd.read_csv(r"/Users/boris/Downloads/БДМО/2 indicator clean v2.csv")
b = BDMO_data_2[["mun_type", "municipality", "oktmo", "oktmo_munr", "rayon", "year"]]

BDMO_data_3 = pd.read_csv(r"/Users/boris/Downloads/БДМО/3 indicator clean v2.csv")
c = BDMO_data_3[["mun_type", "municipality", "oktmo", "oktmo_munr", "rayon", "year"]]
```

```python
# Обьеденяю таблицы a,b,c в одну таблицу 
frames = [a, b, c]
result = pd.concat(frames)
display(result)
```

```python
# создаю новую таблицу с переименоваными колонками 
new = result[['oktmo', 'rayon', "year"]].copy()
new.columns = ['OKTMO', 'MUN_NAME', "YEAR"]
new
```

```python
# Обьединяю BDMO данные с таблицей выше "new"
BDMO_data = pd.read_csv(r"/Users/boris/Downloads/BDMO.csv")
framess = [new,BDMO_data]
resultt = pd.concat(framess)
display(resultt)
```

```python
# убираю дубликаты, убираю из данных слово "муниципальный"
no_dub = resultt.drop_duplicates(subset =["MUN_NAME", 'OKTMO', "YEAR"], keep = "first")
no_dub['MUN_NAME'] = no_dub['MUN_NAME'].str.replace('муниципальный', " ")
no_dub
```

```python
# Соеденяю JSON данные с (BDMO_data + из ссылки в интернете) 
JSON_data = pd.read_csv(r"/Users/boris/Downloads/JSON.csv")
new_test =  JSON_data[['MUN_ID', 'MUN_NAME', 'REGION']].merge(no_dub, how = "left", left_on = 'MUN_NAME',right_on = "MUN_NAME")
new_test
```

```python
# убираю nan
new_test.fillna(0, inplace = True)
new_test[new_test.MUN_NAME != 0]
```

```python
# только 2016 год
new_test_2016 = new_test[new_test.YEAR == 2016]
new_test_2016
```

```python
# удаляю дубликаты из нового обьединеного датасета
new_no_dub = new_test_2016.drop_duplicates(subset =["MUN_NAME", 'OKTMO', "YEAR"], keep = "first")
new_no_dub
```
